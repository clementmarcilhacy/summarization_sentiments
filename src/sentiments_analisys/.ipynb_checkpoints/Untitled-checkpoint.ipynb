```python
import os
import sqlite3
from joblib import load
import pandas as pd
```

```python
def conn():

    try:
        dir_path = os.path.dirname(os.path.abspath('')) + "/mysite/db.sqlite3"
        print(dir_path)
        bbdd = sqlite3.connect(dir_path, timeout=10)
        bbdd.row_factory = sqlite3.Row
        return bbdd.cursor()
        return
    except:
        raise ("Error a conectar a la base de datos.")

    return ""

def queery(cursor):
    cursor.execute("""
                    SELECT *
                    FROM sumtravel_review
                    WHERE local_id = ? and
                    año = ?

                    """, (2,2017))
    rows = cursor.fetchall()
    clf = load('filename.joblib')

    lst = []
    for row in rows:
        dic = {}
        text = row[1]
        dic["frase"] = text
        dic["mes"] = row[2]
        dic["año"] =  row[3]
        dic["sentimiento"] = clf.predict([text])[0]
        lst.append(dic)

    return lst
```

```python
datos = pd.DataFrame(queery(conn()))
```

```python
import spacy
from functools import reduce

nlp = spacy.load('es_core_news_sm')
fdist = frases_positivas.frase.tolist()



suma = reduce(lambda a, x: a + x, fdist)

doc = nlp(suma)

for sent in doc.sents:
    print(sent)
    print("\n")

            
```
